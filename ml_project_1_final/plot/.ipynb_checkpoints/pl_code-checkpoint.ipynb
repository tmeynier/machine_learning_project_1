```python
import pylab as pl
def plotpca(xdata, ydata, target_names, items, filename):
    """Make plot."""
    pca = PCA(n_components=2)
    components = pca.fit(xdata).transform(xdata)

    # Percentage of variance explained for each components
    print('explained variance ratio (first two components):',
        pca.explained_variance_ratio_)

    pl.figure()  # Make a plotting figure
    pl.subplots_adjust(bottom=0.1)

    # NB: a maximum of 7 targets will be plotted
    for i, (c, m, target_name) in enumerate(zip(
            'rbmkycg', 'o^s*v+x', target_names)):
        pl.scatter(components[ydata == i, 0], components[ydata == i, 1],
                color=c, marker=m, label=target_name)
        for n, x, y in zip(
                (ydata == i).nonzero()[0],
                components[ydata == i, 0],
                components[ydata == i, 1]):
            pl.annotate(
                    items[n],
                    xy=(x, y),
                    xytext=(5, 5),
                    textcoords='offset points',
                    color=c,
                    fontsize='small',
                    ha='left',
                    va='top')
    pl.legend()
    pl.title('PCA of %s' % filename)
    pl.show()
```

```python
items = np.arange(X.shape[0])
plotpca(X,Y, np.array(["name1","name2","name3","n4","n5"]),items , "mitbih_train.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

df_train = pd.read_csv("../../mitbih_train.csv", header=None)
df_train = df_train.sample(frac=1)
df_test = pd.read_csv("../../mitbih_test.csv", header=None)

Y = np.array(df_train[187].values).astype(np.int8)
#X = np.array(df_train[list(range(187))].values)[..., np.newaxis]
X = np.array(df_train[list(range(187))].values)[...]
Y_test = np.array(df_test[187].values).astype(np.int8) #Used for tersting
#X_test = np.array(df_test[list(range(187))].values)[..., np.newaxis] #Used for testing 
X_test = np.array(df_test[list(range(187))].values)[...] #Used for testing 
```

```python
x = df_train.loc[: ,range(187)].values
y = df_train.loc[:187].values

x= StandardScaler().fit_transform(x)

vec = y[:,187]
print( vec[vec == 5])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents,
                          columns = ['principal component 1' ,
                                    'principal component 2'])

finalDf = pd.concat([principalDf, df_train[187]] , axis=1)
```

```python
target_dict = {0:'N' , 1:'S' , 2:'V', 3:'F' , 4:'Q'}
target_ind = target_dict.keys()
target_val = target_dict.values()
```

```python
fig = plt.figure(figsize = (8,8))

ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
#targets = ['N': 0, 'S': 1, 'V': 2, 'F': 3, 'Q': 4]
targets = target_ind #range(4)
colors = ['r', 'g', 'b' , 'y' , 'm'] #k = black #m = magenta
for target, color in zip(targets,colors):
    indicesToKeep = finalDf[187] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(target_val)
ax.grid()
```
